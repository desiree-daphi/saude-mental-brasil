```python
import pandas as pd
```

```python
# Carregando os arquivos do DATASUS (O TabNet gera CSV com cabeçalho e notas de rodapé, então foi necessário algumas alterações no código)
df_datasus = pd.read_csv("../data/raw/datasus_internacoes_bruto.csv", sep=";", encoding="latin1",
    skiprows=3,
    skipfooter=12,
    engine="python"
)
```

```python
#Inspeção dos Dados
print("Formato (linhas, colunas):", df_datasus.shape)
print("\nPrimeiras linhas:")
df_datasus.head()
```

```python
#Informações gerais
df_datasus.info()
```

```python
print("Valores nulos por colunas:")
print(df_datasus.isnull().sum())
```

```python
#Limpeza básica
# Renomear colunas para nomes sem espaço e em minúsculas
df_datasus.columns = (df_datasus.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "-")
    .str.replace("a", "a")
    .str.replace("ç", "c")
    .str.replace("é", "e")
    .str.replace("ó", "o")
)

#Removendo linhas onde todos os valores são nulos
df_datasus.dropna(how="all", inplace=True)

#Resultado
print("Colunas após renomeação")
print(df_datasus.columns.tolist())
```

```python
#Salvando arquivo

df_datasus.to_csv("../data/processed/datasus_internacoes_limpo.csv", index=False)
print("Arquivo salvo com sucesso!")
```
